```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['GROQ_API_KEY'] = os.getenv('GROQ_API_KEY')
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(model_provider='groq',
                        model = 'llama-3.3-70b-versatile')

response = model.invoke([{"role": "user", "content": "Hey how are you doing?"}])
print(response.content)
```

Streaming is a process of continous output generation as soon as the LLM generates the response. 

It significantly improves the responses for long content generated from LLM

```python
for chunk in model.stream("Write 2000 words about Langchain"):
    print(chunk.content, end="", flush=True)
```

Batch Processing

Batching is a collection of independent requests to a model can significantly improve performance and reduce cost, as the processing can be done in parallel

```python
responses = model.batch([
    "Why do parrots have colourful feathers?",
    "What is the capital of France?",
    "What is the meaning of life?"],
    config = {
        'max_concurrency': 5
    }
)

for response in responses:
    print(response.content)
```
