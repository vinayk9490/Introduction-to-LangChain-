```python
import os
from langchain.chat_models import init_chat_model

os.environ['GROQ_API_KEY'] = os.getenv('GROQ_API_KEY')
model = init_chat_model(model_provider='groq',
                        model = 'llama-3.3-70b-versatile')
```

TEXT PROMPTS 

Text prompts are the strings - ideal for straightforward generation tasks where you don't need to retain conversation history

```python
model.invoke("What is Langfuse?")
```

Message Prompts 

Message Types:
1. System Message - Tell the model how to behave and provide context for interactions
2. Human Message - Represents user input and interactions with the model
3. AI Message - Response generated by the model, including text content, tool calls and metdata
4. Tool Message - Represents the outputs generated by the tools

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage(content = "You are an intelligent poetry expert."),
    HumanMessage(content = "Write a poem of AI?"),
]

response = model.invoke(messages)
response
print(response.content)
```

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage(content = "You are an intelligent API expert."),
    HumanMessage(content = 'How to create a rest API in Python?'),
]

response = model.invoke(messages)
response
print(response.content)
```

PROVIDING DETAILED INFORMATION TO THE LLM THROUGH SYSTEM PROMPT

```python
from langchain.messages import SystemMessage

system_message = SystemMessage(
    content = "You are an intelligent API Expert. Acts as a senior backend Django Developer and " \
    "Answer the question based on your expertise and experience in backend development and Django framework."
)
Messages = [system_message, HumanMessage(content = "How to create a rest API in Python?")]

result = model.invoke(Messages)
result.content.split("\n")
```

```python
#providing information based on the metadata of the user 

hum_msg = HumanMessage(content = "What is the best way to learn Python?",
                       name="alice",
                       id = "msg_1")
result = model.invoke([hum_msg])
result
```
