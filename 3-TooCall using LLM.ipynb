Tools 

'''
Models can request to call tools that perform tasks such as fetching the data from db, searching from web or running the code. 
'''

```python
import os
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv
load_dotenv()

model = init_chat_model(model_provider='groq',
                        model = 'llama-3.3-70b-versatile')
response = model.invoke("Why do parrots talk?")
print(response.content)
```

```python
from langchain.tools import tool
@tool 
def get_weather(location:str) -> str:
    '''
    Get the current weather for a given location.
    '''
    return f"The current weather in {location} is sunny with a high of 25 degrees Celsius."

model_with_tools = model.bind_tools([get_weather])
response = model_with_tools.invoke("What's the weather like in New York?")
print(response)
print(response.tool_calls)
response.content
```

```python
from langchain.agents import create_agent
from langchain.chat_models import init_chat_model

def get_weather(Location: str)->str:
    '''
    This function takes a location as input and returns a string describing the weather in that location.
    '''
    return f"The weather in {Location} is sunny with a high of 25°C."

#init_chat_model is used to initialize the chat model with the specified model name and provider details. 
model = init_chat_model("llama-3.3-70b-versatile", model_provider="groq")

agent = create_agent(
    model = model,
    tools = [get_weather],
    system_prompt= "You are a helpful assistant that provides weather information based on the location provided by the user."
)

#for create_agent I need to pass the model provider details and the tools that I want to use for create_agent.
```

```python
result = agent.invoke(
    {"messages": [{"role": "user", "content": "What is the weather in New York?"}]},
    context={"user_role": "expert"}
)
print(result['messages'])
print(result['messages'][2].content)
```
